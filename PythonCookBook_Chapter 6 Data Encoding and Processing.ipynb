### 6.1. Reading and Writing CSV Data

```python
import csv
import os
from collections import namedtuple
path = os.getcwd()
with open(path + '\\atp.csv') as f:
    f_csv = csv.reader(f)
    headers = next(f_csv)
    Row = namedtuple('Row', headers)
    for r in f_csv:
        row = Row(*r)
        print(row.winner_name)
```

### 6.3. Parsing Simple XML Data

```python
from urllib.request import urlopen
from xml.etree.cElementTree import parse
u = urlopen('http://planet.python.org/rss20.xml')
doc = parse(u)
print(doc)
for item in doc.iterfind('channel/item'):
    tittle = item.findtext('title')
    date = item.findtext('pubDate')
    link = item.findtext('link')
    
    print(tittle)
    print(date)
    print(link)
    print()
```

```python
from xml.etree.ElementTree import iterparse

def parse_and_remove(filename, path):
    path_parts = path.split('/')
    doc = iterparse(filename, ('start', 'end'))
    next(doc)
    
    tag_stack = []
    elem_stack = []
    for event, elem in doc:
        if event == 'start':
            tag_stack.append(elem.tag)
            elem_stack.append(elem)
        elif event == 'end':
            if tag_stack == path_parts:
                yield elem
                elem_stack[-2].remvoe(elem)
            try:
                tag_stack.pop()
                elem_stack.pop()
            except IndexError:
                pass
```

### 6.5. Turning a Dictionary into XML

```python
from xml.etree.ElementTree import Element
def dict_to_xml(tag, d):
    elem  = Element(tag)
    for key, val in d.items():
        child = Element(key)
        child.text = str(val)
        elem.append(child)
    return elem
ss = {'prod':'chrome'}
ss2 = {'prod':'youtube'}
s = {'name': 'Google', 'shares':100, 'price':490}
exml = dict_to_xml('stock', s)
from xml.etree.ElementTree import tostring
tostring(exml)
```
